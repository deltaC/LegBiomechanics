Гармонические колебания обратного маятника

```python
import numpy as np
import matplotlib.pyplot as plt
import math as m

a = 0.5 #g/l
t_0 = 0
t_finish = 30
n = 2000

t = np.linspace(t_0, t_finish, n)
delta_t = t[2]-t[1]
x_1 = np.zeros(n)
x_2 = np.zeros(n)
x_1[0] = 2
x_2[0] = 0
f1 = []
f2 = []

#f'' - sin(f)*g/l = 0
for i in range(n-1):
    f_1 = x_2[i]
    f_2 = a * m.sin(x_1[i])
    x_1[i+1] = x_1[i] + delta_t * f_1
    x_2[i+1] = x_2[i] + delta_t * f_2

#for j in range(t_0, t_finish):
#    print(f"t = {j:.2f}, x_1 = {x_1[j]:.4f}, x_2 = {x_2[j]:.4f}")

plt.figure(figsize=(8, 6))
plt.plot(t, x_1, label="Угол θ(t)")
plt.plot(t, x_2, label="Угловая скорость ω(t)")
plt.xlabel("Время (с)")
plt.ylabel("Угол (рад) / Угловая скорость (рад/с)")
plt.title("Обратный маятник без трения")
plt.legend()
plt.grid(True)
plt.show()
```

Затухающие колебания обратного маятника

```python
a = 0.6 #g/l
b = 0.09 #k/m
t_0 = 0
t_finish = 40
n = 2000

t = np.linspace(t_0, t_finish, n)
delta_t = t[2]-t[1]
x_1 = np.zeros(n)
x_2 = np.zeros(n)
x_1[0] = 2
x_2[0] = 0
f1 = []
f2 = []

#f'' + f'*k*m - sin(f)*g/l = 0
for i in range(n-1):
    f_1 = x_2[i]
    f_2 = -b * x_2[i] + a * m.sin(x_1[i])
    x_1[i+1] = x_1[i] + delta_t * f_1
    x_2[i+1] = x_2[i] + delta_t * f_2

#for j in range(t_0, t_finish):
#    print(f"t = {j:.2f}, x_1 = {x_1[j]:.4f}, x_2 = {x_2[j]:.4f}")

plt.figure(figsize=(8, 6))
plt.plot(t, x_1, label="Угол θ(t)")
plt.plot(t, x_2, label="Угловая скорость ω(t)")
plt.xlabel("Время (с)")
plt.ylabel("Угол (рад) / Угловая скорость (рад/с)")
plt.title("Обратный маятник c трением")
plt.legend()
plt.grid(True)
plt.show()
```
